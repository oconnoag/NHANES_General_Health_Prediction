# II. Introduction

Assessing subjective feelings (e.g. general health, quality of life, etc) has been an important task in medicine for as long as the profession has existed.  Considering that a core goal in medicine seeks to restore positive subjective perceptions and/or prevent negative subjective perceptions in the subjects, the salience of accurate measurements is not hard to overstate.  As with any subjective assessment, however, a number of biases can introduce a large amount of arbitrary variance to the measurements.  There have been several attempts to standardize measurements of subjective experience using various survey questions--one example is called the Nottingham Health Profile, an attempt to measure "quality of life" (Hunt, 1985).  To expand to the existing list of subjective-to-objective mapping, we are attempting to predict a particular subjective measurement ("General Health") from CDC health data by building statistical models with laboratory-based data, sociological-based data, and combined subsets of the two.  The general hypothesis for this project is given a certain set of information about a patient, we can determine/predict how they feel about their health (by classifying them into a corresponding category), without explicitly asking them.

## II.a Data Description

#### NHANES Overview
The data used in this study is compiled and propogated by the CDC under the program name National Health and Nutrition Examination Survey (which will herein be referred to as NHANES).  The NHANES program operates annually (though with different pools of participants) with the goal of assessing health and nutrition of the denizens of America.  The data produced by the program is unique because it utilizes biochemical laboratory-based methods, questionnaires, diet, external body measures, and demographics.  This abundance of data is attractive for statisticians and data scientists (or students of the fields, like us) for generating models for all sorts of different outcomes using a number of measurements (referred to as features in data science).

#### General Health
The general_health field is originally coded as the "HSD010" column in the /Questionnaire/HSQ_I ("General Health Status") file.  The measurements come from subjects answering the question:  "Would you say {your/SP's} health in general is {List Options}?".  The answers are broken up into 5 levels: 1, 2, 3, 4, 5 corresponding to "Excellent", "Very Good", "Good", "Fair", and "Poor", respectively.  These categories were further mapped to 3 classes, due to poor model performance (likely caused by extreme class inbalances).  The mapping was {classes 1-2: 1, class 3: 2, classes 4-5: 3} and were described as {1: "Better", 2: "Good", 3: "Worse"}.

### Data Cleaning
In order to be able to use our data for model generation (and some visualizations like pairplot), there cannot be nan values in any of the columns.  This led to removing some initial features that didn't have enough datapoints from the start. We also removed any conceptually meaningless values from our sets, such as "Don't know" and "Refused". Biology is messy, so it's difficult to determine what constitutes an outlier; however, we may pick some threshold for distance (in standard deviations from the mean) if certain values seem to be skewing the data in drastic ways (like perhaps glucose measurements).

### Modeling Considerations, Pipeline, and Assessment

We generated a model pipeline to automate data splitting, model construction (with hyperparamter optimization), metric scraping, and performance assessing through plotting (boxplots).  Each model above was constructed 25-times (for each section, respectively) using a different training/test data split on each iteration.  The datasplits were stratified, meaning that each split received the same percentage of each class observation as the full data class-priors.  The hyperparamters were optimized to the dataset using Scikit-learn's GridSearch procedure with 5-fold cross validation to optimize the 'balanced accuracy' of the model.  Balanced accuracy was chosen as the metric, because it is an unweighted assessment of the model's accuracy; this bodes well for the general health data, because the classes are initially unbalanced.  Precision of the models were also assessed to determine if the models were performing well in each category.  It should be noted that a high precision wasn't necessarily favorable (consider, the case were only 10 predictions of class 10 were made and all 10 of those correct; because the precision was macro-averaged, that class has the same weight as the larger classes, so could heavy skew the overall impression of this metric).

Certain alterations to the data/models were also made to assess the performance of the models at their full potential.  For example, KNN was run with both unscaled and scaled data (to account for the 'curse of dimensionality', since several features (particularly in the lab-data) were magnitudes greater than others.  Another consideration for the models that operate through a probability-based classification routine, was changing the threshold for classifying a point as class '3' (worse) to 27%.  The class with the highest predicted percentage is typically given the classification, so this alteration classifies more points to the 3 category.  27% was used as the altered threshold for all of these alterations; however, it should be noted that if a practitioner wanted to use this method in their final models, they could further optimize that percentage.

```python
# Set up metric lists
lab_linreg_bal_acc = []
lab_linreg_prec = []

lab_knn_bal_acc = []
lab_knn_prec = []

lab_knn_sc_bal_acc = []
lab_knn_sc_prec = []

lab_lr_bal_acc = []
lab_lr_prec = []

lab_lr_t_bal_acc = []
lab_lr_t_prec = []

lab_lda_bal_acc = []
lab_lda_prec = []

lab_lda_t_bal_acc = []
lab_lda_t_prec = []

lab_qda_bal_acc = []
lab_qda_prec = []

lab_qda_t_bal_acc = []
lab_qda_t_prec = []

lab_rf_bal_acc = []
lab_rf_prec = []

lab_rf_t_bal_acc = []
lab_rf_t_prec = []

lab_vc_bal_acc = []
lab_vc_prec = []

lab_vc_t_bal_acc = []
lab_vc_t_prec = []

lab_gb_bal_acc = []
lab_gb_prec = []

lab_gb_t_bal_acc = []
lab_gb_t_prec = []

# Number of iterations
cycles = 25;

for _ in range(cycles):
    ###########################################################################################
    ##################### Split Data
    ###########################################################################################
    
    # Stratified training/testing data
    (lab_X_train, lab_X_test, 
    lab_y_train, lab_y_test) = data_splitter(lab_selected_features.drop('general_health', axis=1), 
                                              lab_selected_features['general_health'])
    
    # Scaled & Stratified training/testing data
    (sc_lab_X_train, sc_lab_X_test, 
    sc_lab_y_train, sc_lab_y_test) = scaled_data_splitter(lab_selected_features.drop('general_health', axis=1), 
                                                          lab_selected_features['general_health'])
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Weak Learners ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    
    ###########################################################################################
    ##################### Linear Regression-Based Classification
    ###########################################################################################

    lab_linreg = LinearRegression(n_jobs=-1)
    lab_linreg_m = lab_linreg.fit(lab_X_train, lab_y_train)

    lab_linreg_pred = lab_linreg_m.predict(lab_X_test)
    lab_linreg_pred = lab_linreg_pred.round(0)
    
    lab_linreg_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_linreg_pred) )
    lab_linreg_prec.append( precision_score(lab_y_test, lab_linreg_pred, average='macro') )
    
    ###########################################################################################
    ##################### KNN
    ###########################################################################################
    
    lab_knn = KNeighborsClassifier(n_jobs=-1)

    lab_knn_params = {
        'n_neighbors': [10, 15, 20, 25]
    }

    # Model and append metrics
    lab_knn_ba, lab_knn_p = grid_search_wrap(lab_knn, lab_knn_params, 
                                             lab_X_train, lab_y_train, lab_X_test, lab_y_test)
    lab_knn_bal_acc.append(lab_knn_ba[0])
    lab_knn_prec.append(lab_knn_p)
    
    ###########################################################################################
    ##################### KNN (Scaled)
    ###########################################################################################
    
    lab_sc_knn = KNeighborsClassifier(n_jobs=-1)

    lab_sc_knn_params = {
        'n_neighbors': [10, 15, 20, 25]
    }

    # Model and append metrics
    lab_knn_sc_ba, lab_knn_sc_p = grid_search_wrap(lab_sc_knn, lab_sc_knn_params, 
                                                   sc_lab_X_train, sc_lab_y_train, sc_lab_X_test, sc_lab_y_test)
    lab_knn_sc_bal_acc.append(lab_knn_sc_ba[0])
    lab_knn_sc_prec.append(lab_knn_sc_p)


    ###########################################################################################
    ##################### Logistic Regression
    ###########################################################################################
    
    lab_lr = LogisticRegression(solver='lbfgs', multi_class='multinomial', n_jobs=-1)

    lab_lr_params = {
        'C': [0.1, 1, 10, 100],
    }

    # Model and append metrics
    lab_lr_ba, lab_lr_p = grid_search_wrap(lab_lr, lab_lr_params,
                                           lab_X_train, lab_y_train, lab_X_test, lab_y_test)
    lab_lr_bal_acc.append(lab_lr_ba[0])
    lab_lr_prec.append(lab_lr_p)
    
    ###########################################################################################
    ##################### Logistic Regression (Altered Classification Threshold)
    ###########################################################################################
    
    lab_lr_t = LogisticRegression(solver='lbfgs', multi_class='multinomial', n_jobs=-1)

    lab_lr_t_params = {
    'C': [0.1, 1, 10, 100],
    }
    
    # Model and append metrics
    lab_lr_t_ba, lab_lr_t_p = grid_search_wrap(lab_lr_t, lab_lr_t_params, 
                                               lab_X_train, lab_y_train, lab_X_test, lab_y_test,
                                               altered_threshold=True, threshold=.27)
    lab_lr_t_bal_acc.append(lab_lr_t_ba[0])
    lab_lr_t_prec.append(lab_lr_t_p)

    ###########################################################################################
    ##################### LDA
    ###########################################################################################
    
    lab_lda = LinearDiscriminantAnalysis()

    lab_lda_params = {
        'priors' : [None, lab_priors],
    }

    # Model and append metrics
    lab_lda_ba, lab_lda_p = grid_search_wrap(lab_lda, lab_lda_params, lab_X_train, lab_y_train, lab_X_test, lab_y_test)
    lab_lda_bal_acc.append(lab_lda_ba[0])
    lab_lda_prec.append(lab_lda_p)
    
    ###########################################################################################
    ##################### LDA (Altered Classification Threshold)
    ###########################################################################################
    
    lab_lda_t = LinearDiscriminantAnalysis()

    lab_lda_t_params = {
        'priors' : [None, lab_priors],
    }
    
    # Model and append metrics
    lab_lda_t_ba, lab_lda_t_p = grid_search_wrap(lab_lda_t, lab_lda_t_params, 
                                               lab_X_train, lab_y_train, lab_X_test, lab_y_test,
                                               altered_threshold=True, threshold=.27)
    lab_lda_t_bal_acc.append(lab_lda_t_ba[0])
    lab_lda_t_prec.append(lab_lda_t_p)

    ###########################################################################################
    ##################### QDA
    ###########################################################################################
    
    lab_qda = QuadraticDiscriminantAnalysis()

    lab_qda_params = {
        'priors' : [None, lab_priors],
    }
    
    # Model and append metrics
    lab_qda_ba, lab_qda_p = grid_search_wrap(lab_qda, lab_qda_params, 
                                             lab_X_train, lab_y_train, lab_X_test, lab_y_test)
    lab_qda_bal_acc.append(lab_qda_ba[0])
    lab_qda_prec.append(lab_qda_p)
    
    ###########################################################################################
    ##################### QDA (Altered Classification Threshold)
    ###########################################################################################        
    
    lab_qda_t = QuadraticDiscriminantAnalysis()

    lab_qda_t_params = {
        'priors' : [None, lab_priors],
    }
    
    # Model and append metrics
    lab_qda_t_ba, lab_qda_t_p = grid_search_wrap(lab_qda_t, lab_qda_t_params, 
                                                 lab_X_train, lab_y_train, lab_X_test, lab_y_test)
    lab_qda_t_bal_acc.append(lab_qda_t_ba[0])
    lab_qda_t_prec.append(lab_qda_t_p)
   

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Ensemble Methods ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ###########################################################################################
    ##################### Random Forest (Normal/Altered Threshold)
    ###########################################################################################
    # These hyperparameters were selected using GridSearching outside of this loop -- just too 
    # slow to have run many times
    lab_rf = RandomForestClassifier(n_estimators=50, max_depth=100, max_features=2, n_jobs=-1)
    lab_rf_m = lab_rf.fit(lab_X_train, lab_y_train)
    
    # Normal Predicting / Metrics
    lab_rf_pred = lab_rf.predict(lab_X_test)
    lab_rf_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_rf_pred) )
    lab_rf_prec.append( precision_score(lab_y_test, lab_rf_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    lab_rf_t_pred = predict_w_treshold(lab_rf, lab_X_test, class_num=3, threshold=0.27)
    lab_rf_t_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_rf_t_pred) )
    lab_rf_t_prec.append( precision_score(lab_y_test, lab_rf_t_pred, average='macro') )
    
    ###########################################################################################
    ##################### Voting Classifier (Soft Voting) (Normal/Altered Threshold)
    ###########################################################################################
    # These Hyperparamters were chosen from GridSearching the individual models above
    clf1 = sklearn.ensemble.RandomForestClassifier(n_estimators=50)
    clf2 = sklearn.discriminant_analysis.LinearDiscriminantAnalysis(priors=lab_priors)
    clf3 = sklearn.linear_model.LogisticRegressionCV()
    clf4 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=25)

    lab_vc = sklearn.ensemble.VotingClassifier(estimators=[('rf', clf1), ('lda', clf2), ('lr', clf3), ('knn', clf4)],
                                              n_jobs=-1, voting='soft')

    lab_vc_model = lab_vc.fit(lab_X_train, lab_y_train)
    
    # Normal Predicting / Metrics
    lab_vc_pred = lab_vc.predict(lab_X_test)
    lab_vc_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_vc_pred) )
    lab_vc_prec.append( precision_score(lab_y_test, lab_vc_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    lab_vc_t_pred = predict_w_treshold(lab_vc, lab_X_test, class_num=3, threshold=0.27)
    lab_vc_t_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_vc_t_pred) )
    lab_vc_t_prec.append( precision_score(lab_y_test, lab_vc_t_pred, average='macro') )    
    
    ###########################################################################################
    ##################### Gradient Boosting (Normal/Altered Threshold)
    ###########################################################################################
    # These hyperparameters were selected using GridSearching outside of this loop -- just too 
    # slow to have run many times
    lab_gb = GradientBoostingClassifier(max_features=2, n_estimators=50)
    
    lab_gb_m = lab_gb.fit(lab_X_train, lab_y_train)
    
    # Normal Predicting / Metrics
    lab_gb_pred = lab_gb.predict(lab_X_test)
    lab_gb_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_gb_pred) )
    lab_gb_prec.append( precision_score(lab_y_test, lab_gb_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    lab_gb_t_pred = predict_w_treshold(lab_gb, lab_X_test, class_num=3, threshold=0.27)
    lab_gb_t_bal_acc.append( balanced_accuracy_score(lab_y_test, lab_gb_t_pred) )
    lab_gb_t_prec.append( precision_score(lab_y_test, lab_gb_t_pred, average='macro') ) 
    
    ###########################################################################################
    ##################### Print Status
    ###########################################################################################
    clear_output() # Ensures only one line of text is printed at once
    display( str(_+1) + " Out of " + str(cycles) + " completed . . .")

clear_output()
print("Modeling Complete.")
```

#### Model Building

```python
# Set up metric lists
socio_linreg_bal_acc = []
socio_linreg_prec = []

socio_knn_bal_acc = []
socio_knn_prec = []

socio_knn_sc_bal_acc = []
socio_knn_sc_prec = []

socio_lr_bal_acc = []
socio_lr_prec = []

socio_lr_t_bal_acc = []
socio_lr_t_prec = []

socio_lda_bal_acc = []
socio_lda_prec = []

socio_lda_t_bal_acc = []
socio_lda_t_prec = []

socio_qda_bal_acc = []
socio_qda_prec = []

socio_qda_t_bal_acc = []
socio_qda_t_prec = []

socio_rf_bal_acc = []
socio_rf_prec = []

socio_rf_t_bal_acc = []
socio_rf_t_prec = []

socio_vc_bal_acc = []
socio_vc_prec = []

socio_vc_t_bal_acc = []
socio_vc_t_prec = []

socio_gb_bal_acc = []
socio_gb_prec = []

socio_gb_t_bal_acc = []
socio_gb_t_prec = []

# Number of iterations
cycles = 25;

for _ in range(cycles):
    ###########################################################################################
    ##################### Split Data
    ###########################################################################################
    
    # Stratified training/testing data
    (socio_X_train, socio_X_test, 
    socio_y_train, socio_y_test) = data_splitter(selected_features.drop('general_health', axis=1), 
                                              selected_features['general_health'])
    
    # Scaled & Stratified training/testing data
    (sc_socio_X_train, sc_socio_X_test, 
    sc_socio_y_train, sc_socio_y_test) = scaled_data_splitter(selected_features.drop('general_health', axis=1), 
                                                          selected_features['general_health'])
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Weak Learners ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    
    ###########################################################################################
    ##################### Linear Regression-Based Classification
    ###########################################################################################

    socio_linreg = LinearRegression(n_jobs=-1)
    socio_linreg_m = socio_linreg.fit(socio_X_train, socio_y_train)

    socio_linreg_pred = socio_linreg_m.predict(socio_X_test)
    socio_linreg_pred = socio_linreg_pred.round(0)
    
    socio_linreg_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_linreg_pred) )
    socio_linreg_prec.append( precision_score(socio_y_test, socio_linreg_pred, average='macro') )
    
    ###########################################################################################
    ##################### KNN
    ###########################################################################################
    
    socio_knn = KNeighborsClassifier(n_jobs=-1)

    socio_knn_params = {
        'n_neighbors': [10, 15, 20, 25]
    }

    # Model and append metrics
    socio_knn_ba, socio_knn_p = grid_search_wrap(socio_knn, socio_knn_params, 
                                             socio_X_train, socio_y_train, socio_X_test, socio_y_test)
    socio_knn_bal_acc.append(socio_knn_ba[0])
    socio_knn_prec.append(socio_knn_p)
    
    ###########################################################################################
    ##################### KNN (Scaled)
    ###########################################################################################
    
    socio_sc_knn = KNeighborsClassifier(n_jobs=-1)

    socio_sc_knn_params = {
        'n_neighbors': [10, 15, 20, 25]
    }

    # Model and append metrics
    socio_knn_sc_ba, socio_knn_sc_p = grid_search_wrap(socio_sc_knn, socio_sc_knn_params, 
                                                   sc_socio_X_train, sc_socio_y_train, sc_socio_X_test, sc_socio_y_test)
    socio_knn_sc_bal_acc.append(socio_knn_sc_ba[0])
    socio_knn_sc_prec.append(socio_knn_sc_p)


    ###########################################################################################
    ##################### Logistic Regression
    ###########################################################################################
    
    socio_lr = LogisticRegression(solver='lbfgs', multi_class='multinomial', n_jobs=-1)

    socio_lr_params = {
        'C': [0.1, 1, 10, 100],
    }

    # Model and append metrics
    socio_lr_ba, socio_lr_p = grid_search_wrap(socio_lr, socio_lr_params,
                                           socio_X_train, socio_y_train, socio_X_test, socio_y_test)
    socio_lr_bal_acc.append(socio_lr_ba[0])
    socio_lr_prec.append(socio_lr_p)
    
    ###########################################################################################
    ##################### Logistic Regression (Altered Classification Threshold)
    ###########################################################################################
    
    socio_lr_t = LogisticRegression(solver='lbfgs', multi_class='multinomial', n_jobs=-1)

    socio_lr_t_params = {
    'C': [0.1, 1, 10, 100],
    }
    
    # Model and append metrics
    socio_lr_t_ba, socio_lr_t_p = grid_search_wrap(socio_lr_t, socio_lr_t_params, 
                                               socio_X_train, socio_y_train, socio_X_test, socio_y_test,
                                               altered_threshold=True, threshold=.27)
    socio_lr_t_bal_acc.append(socio_lr_t_ba[0])
    socio_lr_t_prec.append(socio_lr_t_p)

    ###########################################################################################
    ##################### LDA
    ###########################################################################################
    
    socio_lda = LinearDiscriminantAnalysis()

    socio_lda_params = {
        'priors' : [None, socio_priors],
    }

    # Model and append metrics
    socio_lda_ba, socio_lda_p = grid_search_wrap(socio_lda, socio_lda_params, socio_X_train, socio_y_train, socio_X_test, socio_y_test)
    socio_lda_bal_acc.append(socio_lda_ba[0])
    socio_lda_prec.append(socio_lda_p)
    
    ###########################################################################################
    ##################### LDA (Altered Classification Threshold)
    ###########################################################################################
    
    socio_lda_t = LinearDiscriminantAnalysis()

    socio_lda_t_params = {
        'priors' : [None, socio_priors],
    }
    
    # Model and append metrics
    socio_lda_t_ba, socio_lda_t_p = grid_search_wrap(socio_lda_t, socio_lda_t_params, 
                                               socio_X_train, socio_y_train, socio_X_test, socio_y_test,
                                               altered_threshold=True, threshold=.27)
    socio_lda_t_bal_acc.append(socio_lda_t_ba[0])
    socio_lda_t_prec.append(socio_lda_t_p)

    ###########################################################################################
    ##################### QDA
    ###########################################################################################
    
    socio_qda = QuadraticDiscriminantAnalysis()

    socio_qda_params = {
        'priors' : [None, socio_priors],
    }
    
    # Model and append metrics
    socio_qda_ba, socio_qda_p = grid_search_wrap(socio_qda, socio_qda_params, 
                                             socio_X_train, socio_y_train, socio_X_test, socio_y_test)
    socio_qda_bal_acc.append(socio_qda_ba[0])
    socio_qda_prec.append(socio_qda_p)
    
    ###########################################################################################
    ##################### QDA (Altered Classification Threshold)
    ###########################################################################################        
    
    socio_qda_t = QuadraticDiscriminantAnalysis()

    socio_qda_t_params = {
        'priors' : [None, socio_priors],
    }
    
    # Model and append metrics
    socio_qda_t_ba, socio_qda_t_p = grid_search_wrap(socio_qda_t, socio_qda_t_params, 
                                                 socio_X_train, socio_y_train, socio_X_test, socio_y_test)
    socio_qda_t_bal_acc.append(socio_qda_t_ba[0])
    socio_qda_t_prec.append(socio_qda_t_p)
   

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Ensemble Methods ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ###########################################################################################
    ##################### Random Forest (Normal/Altered Threshold)
    ###########################################################################################
    # These hyperparameters were selected using GridSearching outside of this loop -- just too 
    # slow to have run many times
    socio_rf = RandomForestClassifier(n_estimators=50, max_depth=100, max_features=2, n_jobs=-1)
    socio_rf_m = socio_rf.fit(socio_X_train, socio_y_train)
    
    # Normal Predicting / Metrics
    socio_rf_pred = socio_rf.predict(socio_X_test)
    socio_rf_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_rf_pred) )
    socio_rf_prec.append( precision_score(socio_y_test, socio_rf_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    socio_rf_t_pred = predict_w_treshold(socio_rf, socio_X_test, class_num=3, threshold=0.27)
    socio_rf_t_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_rf_t_pred) )
    socio_rf_t_prec.append( precision_score(socio_y_test, socio_rf_t_pred, average='macro') )
    
    ###########################################################################################
    ##################### Voting Classifier (Soft Voting) (Normal/Altered Threshold)
    ###########################################################################################
    # These Hyperparamters were chosen from GridSearching the individual models above
    clf1 = sklearn.ensemble.RandomForestClassifier(n_estimators=50)
    clf2 = sklearn.discriminant_analysis.LinearDiscriminantAnalysis(priors=socio_priors)
    clf3 = sklearn.linear_model.LogisticRegressionCV()
    clf4 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=25)

    socio_vc = sklearn.ensemble.VotingClassifier(estimators=[('rf', clf1), ('lda', clf2), ('lr', clf3), ('knn', clf4)],
                                              n_jobs=-1, voting='soft')

    socio_vc_model = socio_vc.fit(socio_X_train, socio_y_train)
    
    # Normal Predicting / Metrics
    socio_vc_pred = socio_vc.predict(socio_X_test)
    socio_vc_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_vc_pred) )
    socio_vc_prec.append( precision_score(socio_y_test, socio_vc_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    socio_vc_t_pred = predict_w_treshold(socio_vc, socio_X_test, class_num=3, threshold=0.27)
    socio_vc_t_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_vc_t_pred) )
    socio_vc_t_prec.append( precision_score(socio_y_test, socio_vc_t_pred, average='macro') )    
    
    ###########################################################################################
    ##################### Gradient Boosting (Normal/Altered Threshold)
    ###########################################################################################
    # These hyperparameters were selected using GridSearching outside of this loop -- just too 
    # slow to have run many times
    socio_gb = GradientBoostingClassifier(max_features=2, n_estimators=50)
    
    socio_gb_m = socio_gb.fit(socio_X_train, socio_y_train)
    
    # Normal Predicting / Metrics
    socio_gb_pred = socio_gb.predict(socio_X_test)
    socio_gb_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_gb_pred) )
    socio_gb_prec.append( precision_score(socio_y_test, socio_gb_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    socio_gb_t_pred = predict_w_treshold(socio_gb, socio_X_test, class_num=3, threshold=0.27)
    socio_gb_t_bal_acc.append( balanced_accuracy_score(socio_y_test, socio_gb_t_pred) )
    socio_gb_t_prec.append( precision_score(socio_y_test, socio_gb_t_pred, average='macro') ) 
    
    ###########################################################################################
    ##################### Print Status
    ###########################################################################################
    clear_output() # Ensures only one line of text is printed at once
    display( str(_+1) + " Out of " + str(cycles) + " completed . . .")

clear_output()
print("Modeling Complete.")
```

```python
# Set up metric lists
combined_linreg_bal_acc = []
combined_linreg_prec = []

combined_knn_bal_acc = []
combined_knn_prec = []

combined_knn_sc_bal_acc = []
combined_knn_sc_prec = []

combined_lr_bal_acc = []
combined_lr_prec = []

combined_lr_t_bal_acc = []
combined_lr_t_prec = []

combined_lda_bal_acc = []
combined_lda_prec = []

combined_lda_t_bal_acc = []
combined_lda_t_prec = []

combined_qda_bal_acc = []
combined_qda_prec = []

combined_qda_t_bal_acc = []
combined_qda_t_prec = []

combined_rf_bal_acc = []
combined_rf_prec = []

combined_rf_t_bal_acc = []
combined_rf_t_prec = []

combined_vc_bal_acc = []
combined_vc_prec = []

combined_vc_t_bal_acc = []
combined_vc_t_prec = []

combined_gb_bal_acc = []
combined_gb_prec = []

combined_gb_t_bal_acc = []
combined_gb_t_prec = []

# Number of iterations
cycles = 25;

for _ in range(cycles):
    ###########################################################################################
    ##################### Split Data
    ###########################################################################################
    
    # Stratified training/testing data
    (combined_X_train, combined_X_test, 
    combined_y_train, combined_y_test) = data_splitter(combined_selected_features.drop('general_health', axis=1), 
                                              combined_selected_features['general_health'])
    
    # Scaled & Stratified training/testing data
    (sc_combined_X_train, sc_combined_X_test, 
    sc_combined_y_train, sc_combined_y_test) = scaled_data_splitter(combined_selected_features.drop('general_health', axis=1), 
                                                          combined_selected_features['general_health'])
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Weak Learners ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    
    ###########################################################################################
    ##################### Linear Regression-Based Classification
    ###########################################################################################

    combined_linreg = LinearRegression(n_jobs=-1)
    combined_linreg_m = combined_linreg.fit(combined_X_train, combined_y_train)

    combined_linreg_pred = combined_linreg_m.predict(combined_X_test)
    combined_linreg_pred = combined_linreg_pred.round(0)
    
    combined_linreg_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_linreg_pred) )
    combined_linreg_prec.append( precision_score(combined_y_test, combined_linreg_pred, average='macro') )
    
    ###########################################################################################
    ##################### KNN
    ###########################################################################################
    
    combined_knn = KNeighborsClassifier(n_jobs=-1)

    combined_knn_params = {
        'n_neighbors': [10, 15, 20, 25]
    }

    # Model and append metrics
    combined_knn_ba, combined_knn_p = grid_search_wrap(combined_knn, combined_knn_params, 
                                             combined_X_train, combined_y_train, combined_X_test, combined_y_test)
    combined_knn_bal_acc.append(combined_knn_ba[0])
    combined_knn_prec.append(combined_knn_p)
    
    ###########################################################################################
    ##################### KNN (Scaled)
    ###########################################################################################
    
    combined_sc_knn = KNeighborsClassifier(n_jobs=-1)

    combined_sc_knn_params = {
        'n_neighbors': [10, 15, 20, 25]
    }

    # Model and append metrics
    combined_knn_sc_ba, combined_knn_sc_p = grid_search_wrap(combined_sc_knn, combined_sc_knn_params, 
                                                   sc_combined_X_train, sc_combined_y_train, sc_combined_X_test, sc_combined_y_test)
    combined_knn_sc_bal_acc.append(combined_knn_sc_ba[0])
    combined_knn_sc_prec.append(combined_knn_sc_p)


    ###########################################################################################
    ##################### Logistic Regression
    ###########################################################################################
    
    combined_lr = LogisticRegression(solver='lbfgs', multi_class='multinomial', n_jobs=-1)

    combined_lr_params = {
        'C': [0.1, 1, 10, 100],
    }

    # Model and append metrics
    combined_lr_ba, combined_lr_p = grid_search_wrap(combined_lr, combined_lr_params,
                                           combined_X_train, combined_y_train, combined_X_test, combined_y_test)
    combined_lr_bal_acc.append(combined_lr_ba[0])
    combined_lr_prec.append(combined_lr_p)
    
    ###########################################################################################
    ##################### Logistic Regression (Altered Classification Threshold)
    ###########################################################################################
    
    combined_lr_t = LogisticRegression(solver='lbfgs', multi_class='multinomial', n_jobs=-1)

    combined_lr_t_params = {
    'C': [0.1, 1, 10, 100],
    }
    
    # Model and append metrics
    combined_lr_t_ba, combined_lr_t_p = grid_search_wrap(combined_lr_t, combined_lr_t_params, 
                                               combined_X_train, combined_y_train, combined_X_test, combined_y_test,
                                               altered_threshold=True, threshold=.27)
    combined_lr_t_bal_acc.append(combined_lr_t_ba[0])
    combined_lr_t_prec.append(combined_lr_t_p)

    ###########################################################################################
    ##################### LDA
    ###########################################################################################
    
    combined_lda = LinearDiscriminantAnalysis()

    combined_lda_params = {
        'priors' : [None, combined_priors],
    }

    # Model and append metrics
    combined_lda_ba, combined_lda_p = grid_search_wrap(combined_lda, combined_lda_params, combined_X_train, combined_y_train, combined_X_test, combined_y_test)
    combined_lda_bal_acc.append(combined_lda_ba[0])
    combined_lda_prec.append(combined_lda_p)
    
    ###########################################################################################
    ##################### LDA (Altered Classification Threshold)
    ###########################################################################################
    
    combined_lda_t = LinearDiscriminantAnalysis()

    combined_lda_t_params = {
        'priors' : [None, combined_priors],
    }
    
    # Model and append metrics
    combined_lda_t_ba, combined_lda_t_p = grid_search_wrap(combined_lda_t, combined_lda_t_params, 
                                               combined_X_train, combined_y_train, combined_X_test, combined_y_test,
                                               altered_threshold=True, threshold=.27)
    combined_lda_t_bal_acc.append(combined_lda_t_ba[0])
    combined_lda_t_prec.append(combined_lda_t_p)

    ###########################################################################################
    ##################### QDA
    ###########################################################################################
    
    combined_qda = QuadraticDiscriminantAnalysis()

    combined_qda_params = {
        'priors' : [None, combined_priors],
    }
    
    # Model and append metrics
    combined_qda_ba, combined_qda_p = grid_search_wrap(combined_qda, combined_qda_params, 
                                             combined_X_train, combined_y_train, combined_X_test, combined_y_test)
    combined_qda_bal_acc.append(combined_qda_ba[0])
    combined_qda_prec.append(combined_qda_p)
    
    ###########################################################################################
    ##################### QDA (Altered Classification Threshold)
    ###########################################################################################        
    
    combined_qda_t = QuadraticDiscriminantAnalysis()

    combined_qda_t_params = {
        'priors' : [None, combined_priors],
    }
    
    # Model and append metrics
    combined_qda_t_ba, combined_qda_t_p = grid_search_wrap(combined_qda_t, combined_qda_t_params, 
                                                 combined_X_train, combined_y_train, combined_X_test, combined_y_test)
    combined_qda_t_bal_acc.append(combined_qda_t_ba[0])
    combined_qda_t_prec.append(combined_qda_t_p)
   

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Ensemble Methods ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ###########################################################################################
    ##################### Random Forest (Normal/Altered Threshold)
    ###########################################################################################
    # These hyperparameters were selected using GridSearching outside of this loop -- just too 
    # slow to have run many times
    combined_rf = RandomForestClassifier(n_estimators=50, max_depth=100, max_features=2, n_jobs=-1)
    combined_rf_m = combined_rf.fit(combined_X_train, combined_y_train)
    
    # Normal Predicting / Metrics
    combined_rf_pred = combined_rf.predict(combined_X_test)
    combined_rf_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_rf_pred) )
    combined_rf_prec.append( precision_score(combined_y_test, combined_rf_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    combined_rf_t_pred = predict_w_treshold(combined_rf, combined_X_test, class_num=3, threshold=0.27)
    combined_rf_t_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_rf_t_pred) )
    combined_rf_t_prec.append( precision_score(combined_y_test, combined_rf_t_pred, average='macro') )
    
    ###########################################################################################
    ##################### Voting Classifier (Soft Voting) (Normal/Altered Threshold)
    ###########################################################################################
    # These Hyperparamters were chosen from GridSearching the individual models above
    clf1 = sklearn.ensemble.RandomForestClassifier(n_estimators=50)
    clf2 = sklearn.discriminant_analysis.LinearDiscriminantAnalysis(priors=combined_priors)
    clf3 = sklearn.linear_model.LogisticRegressionCV()
    clf4 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=25)

    combined_vc = sklearn.ensemble.VotingClassifier(estimators=[('rf', clf1), ('lda', clf2), ('lr', clf3), ('knn', clf4)],
                                              n_jobs=-1, voting='soft')

    combined_vc_model = combined_vc.fit(combined_X_train, combined_y_train)
    
    # Normal Predicting / Metrics
    combined_vc_pred = combined_vc.predict(combined_X_test)
    combined_vc_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_vc_pred) )
    combined_vc_prec.append( precision_score(combined_y_test, combined_vc_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    combined_vc_t_pred = predict_w_treshold(combined_vc, combined_X_test, class_num=3, threshold=0.27)
    combined_vc_t_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_vc_t_pred) )
    combined_vc_t_prec.append( precision_score(combined_y_test, combined_vc_t_pred, average='macro') )    
    
    ###########################################################################################
    ##################### Gradient Boosting (Normal/Altered Threshold)
    ###########################################################################################
    # These hyperparameters were selected using GridSearching outside of this loop -- just too 
    # slow to have run many times
    combined_gb = GradientBoostingClassifier(max_features=2, n_estimators=50)
    
    combined_gb_m = combined_gb.fit(combined_X_train, combined_y_train)
    
    # Normal Predicting / Metrics
    combined_gb_pred = combined_gb.predict(combined_X_test)
    combined_gb_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_gb_pred) )
    combined_gb_prec.append( precision_score(combined_y_test, combined_gb_pred, average='macro') )
    
    # Threshold Altered Predicting / Metrics
    combined_gb_t_pred = predict_w_treshold(combined_gb, combined_X_test, class_num=3, threshold=0.27)
    combined_gb_t_bal_acc.append( balanced_accuracy_score(combined_y_test, combined_gb_t_pred) )
    combined_gb_t_prec.append( precision_score(combined_y_test, combined_gb_t_pred, average='macro') ) 
    
    ###########################################################################################
    ##################### Print Status
    ###########################################################################################
    clear_output() # Ensures only one line of text is printed at once
    display( str(_+1) + " Out of " + str(cycles) + " completed . . .")

clear_output()
print("Modeling Complete.")
```
